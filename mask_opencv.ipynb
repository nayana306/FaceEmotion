```python
import matplotlib.pylab as plt
import cv2
import numpy as np
```

```python
image = cv2.imread('/Users/nayana/Desktop/test_image.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
print(image.shape) #find the shape of the image
height = image.shape[0] #defining height and width of image
width = image.shape[1]
```

```python
#define region of intrest
region_of_interest_vertices = [
    (0, height),
    (width/2, height/2),
    (width, height)
]
```

```python
def region_of_interest(img, vertices):
    mask = np.zeros_like(img)
    channel_count = img.shape[2]
    match_mask_color = (255,) * channel_count
    cv2.fillPoly(mask, vertices, match_mask_color) #masked all other things than roi
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
```

```python

cropped_image = region_of_interest(image,
                np.array([region_of_interest_vertices], np.int32),)
```

```python
plt.imshow(cropped_image)
plt.show()
```
